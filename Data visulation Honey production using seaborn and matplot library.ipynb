# 3. Display the percentage distribution of the data in each year using the pie chart.

```python
df.groupby('year').count()[['state']].plot(y='state',kind='pie',figsize=(6,6),autopct='%1.1f%%', legend = False)
```

# 4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings

```python
sns.displot(x=df['priceperlb'], kind= 'hist',bins = 30, kde = True)
plt.show()


# using Histogram,plot for ['priceperlb'] and the distribution of the graph is peak in somepoint.
# where the Priceperlb value in between 1.3 and 1.5 and count in between 50 to 60.
# so the graph is positively skewed.
```

# 5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings

```python
# plot the relationship b/w the varibles 'numcol' and 'prodval' through scatterplot.

plt.scatter(x=df['numcol'], y=df['prodvalue'])
plt.title('numcol vs prodvalue')
plt.show()

# The Relationship between 'numcol' and 'prodvalue' is directly related with the representation of graph'
# The values of numcol and prodvalue is merged in
# The Graph values is aslo gradually  increasing
```

# 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot,and write your findings.


```python
sns.catplot(data=df, x='year', y='prodvalue', kind='box', height=5, aspect =1.2)
plt.show()

# Relation between the 'year' and 'prodvalue'is find using seaborn as sns
# Many outliers as obtain for data by using boxplot
```

# 7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inference .(use columns 'numcol','yield percol',total prod,'prodvalue','year').

```python
sns.pairplot(df[['numcol','yieldpercol','totalprod','prodvalue','year']],corner = True)


# 'Numcol','Totalprod','Prodvalue' ard positvely skewed
# 'Yieldpercol' are seems to be Normal Distribution graph.
# so the 'numcol','totalprod','Yieldpercol','prodvalue' seems to be positively related to each other .
```
